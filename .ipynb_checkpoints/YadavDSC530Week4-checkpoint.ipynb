```python
# Lets plot the basic histogram for the field
kids = thinkstats2.Hist(resp.numkdhh, label="Num of Children under 18")
thinkplot.Hist(kids)
thinkplot.Config(xlabel="Num of kids under 18 in a household", ylabel="Count")
```

```python
#Create PMF 
pmfkids = thinkstats2.Pmf(resp.numkdhh, label="Num of Children under 18")
```

```python
# Plot PMF
thinkplot.Hist(pmfkids)
thinkplot.Config(xlabel="Num of kids under 18 in a household", ylabel="PMF")
```

```python
# Biased pmf : the Class Size Paradox

# Use existing function in solution 
def BiasPmf(pmf, label):
    new_pmf = pmf.Copy(label=label)

    for x, p in pmf.Items():
        new_pmf.Mult(x, x)

    new_pmf.Normalize()
    return new_pmf

# apply function
biasedpmfkids =  BiasPmf(pmfkids, label="KidsView")
biasedpmfkids
```

```python
# Plot Biased PMF
thinkplot.Hist(biasedpmfkids)
thinkplot.Config(xlabel="Num of kids under 18 in a household", ylabel="PMF")
```

```python
#Plot the diff between biased and unbiased PMF's, The class size paradox
thinkplot.Pmfs([pmfkids, biasedpmfkids])
thinkplot.Config(xlabel="Num of kids under 18 in a household", ylabel="PMF")
```

```python
def PmfMean(pmf):
    """Computes the mean of a PMF.

    Returns:
        float mean
    """
    mean = 0.0
    for x, p in pmf.d.items():
        mean += p * x
    return mean
```

```python
def PmfVar(pmf, mu=None):
    """Computes the variance of a PMF.

    Args:
        mu: the point around which the variance is computed;
            if omitted, computes the mean

    Returns:
        float variance
    """
    if mu is None:
        mu = pmf.Mean()

    var = 0.0
    for x, p in pmf.d.items():
        var += p * (x - mu) ** 2
    return var
```

```python
# Compute the mean for biased pmf and compare with mean.
PmfMean(biasedpmfkids)
```
